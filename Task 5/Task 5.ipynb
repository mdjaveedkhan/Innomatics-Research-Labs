## Problem 1: Social Media – Post Engagement Analyzer

```python
# Function to calculate total likes and check status
def analyze_post(likes):
    total = 0
    # Add all likes using loop
    for l in likes:
        total += l
    
    print("Total Likes:", total)
    # if a post get likes morethan 1000 then the post is viral
    if total >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

likes = [200, 350, 500]
analyze_post(likes)
```

## Problem 3: Agriculture – Rainfall Adequacy Checker

```python
#  function to analyzes seasonal rainfall data
def check_rainfall(data, required):
    total = 0
    for rain in data:
        total += rain
    # Calculate average rainfall
    avg = total / len(data)
    print("Average Rainfall:", int(avg))
    # if the avg is greaterthan required level then its adequate rainfall
    if avg >= required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

rainfall = [60, 75, 80]
check_rainfall(rainfall, 70)
```

## Problem 4: Social Media – Duplicate Account Detection

```python
# Function to detect duplicate usernames
def detect_duplicates(usernames):
    # Converting list to set
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users = ["user1", "user2", "user1"]
detect_duplicates(users)
```

## Problem 6: Agriculture – Premium Crop Price Filter

## Problem 7: System Monitoring – Application Health Checker

## Problem 9: E-Learning – Student Attendance Eligibility System

```python
# Function to calculate attendance percentage and check exam eligibility
def check_attendance(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1
    # converting into percentage         
    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

attendance = ["P", "P", "A", "P", "P"]
check_attendance(attendance)
```

## Problem 10: Smart Electricity Bill Analyzer

```python
# Function to calculates the electricity bill and classifies usage level based on units consumed
def calculate_bill(units):
    bill = 0
    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    return bill, status

bill, status = calculate_bill(250)
print("Total Bill:", bill)
print("Usage Level:", status)
```
