## Problem 1: Mobile Recharge Validation System

```python
# Function to validate recharge
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    while True:   # loop allows retry
        amount = int(input("Enter recharge amount: "))
        # Check minimum amount
        if amount <=50:
            print("Invalid: Minimum recharge is ₹50")
            continue
        # Check if plan exists
        if amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid Plan. Please try again.")

# Call function
validate_recharge()
```

## Problem 2: Inventory Reorder Alert System

```python
# Function to check stock status
def check_inventory(products):
    # Loop through dictionary
    for item, stock in products.items():
        if stock < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")

# Input data
products = {
    "Rice": 10,
    "Oil": 25,
    "Sugar": 8
}
check_inventory(products)
```

## Problem 3: Student Result Processing System

```python
# Function to process results
def check_result(marks):
    total = 0
    # Calculate total marks
    for i in marks:
        total += i
    average = total / len(marks)
    print("Average Marks:", average)
    
    # Check pass or fail
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [60, 45, 70, 55]
print("Result:", check_result(marks))
```

## Problem 4: Cab Fare Estimator with Retry Option

```python
# Function to calculate fare price
def calculate_fare(distance, peak):
    fare = 50 + (distance * 12) 
    # Add peak hour charge
    if peak == "yes":
        fare = fare + (fare * 0.25)
    return fare
    
# Retry loop
while True:
    km = int(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ").lower()
    # calculate total fare price
    total_fare = calculate_fare(km, peak_hour)
    print("Total Fare:", total_fare)
    retry = input("Do you want to calculate again? (yes/no): ")
    # break if retry is no
    if retry == "no":
        break
```

## Problem 5: Employee Attendance Eligibility Checker

```python
# Function to check eligibility
def check_attendance(attendance):
    present = 0
    # Count present days
    for day in attendance:
        if day == "P":
            present += 1
    # calculate percentage        
    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "A"]
print(check_attendance(attendance))
```

## Problem 6: Password Strength Checker

```python
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False
    
    # Check each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    # Apply rules
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
print(check_password(pwd))
```
