## Problem 1: Unique Words in a Sentence

```python
#take input from the user
sentence = input("Enter sentence: ")

# Split sentence into words using space
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

## Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Initialize variables to track the highest values
max_salary = 0
highest_employee = None

# Iterate through the dictionary items
for name, salary in employees.items():
    if salary > max_salary:
        max_salary = salary
        highest_employee = name

print(f"Highest Salary: {highest_employee} - {max_salary}")
```

## Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

# Assuming first number from the list as both max and min
maximum = numbers[0]
minimum = numbers[0]

# Compare each number in list
for num in numbers:
    #if num is greaterthan maximum then num will be the maximum
    if num > maximum:
        maximum = num
    #if num is lessthan minimmum then num will  be the minimum  
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

## Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
#Set initial count with zero
count = 0

# Check each price
for price in prices:
    #if price is greaterthan 1000 then count increase by 1
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

## Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

#present count initial set to zero
present_count = 0

# Count number of Present days
for day in attendance:
    #if the day is present then count increase by 1
    if day == "P":
        present_count += 1
#calculating the percentage
percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

## Problem 7: Count Character Frequency

```python
text = "python"

#initialising empty dict to store/count the frequency 
frequency = {}

# Loop through each character
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("Character Frequency:", frequency)
```

## Problem 9: Check if a Key Exists in a Dictionary

```python
employee= {"Ravi": 75000, "Anita": 68000, "Kiran": 72000}

#set key as ravi
key = "Ravi"

# Check key existence using 'in'
if key in employee:
    print("Employee exists")
else:
    print("Employee not found")
```

## Problem 10: Calculate Average Marks

```python
marks = [70, 75, 80, 65]

# Calculate sum of marks
total = sum(marks)

# Calculate average
average = total / len(marks)

print("Average Marks:", average)
```
