## Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find employees with high score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

## Problem Statement 2: Search Query Keyword Analysis

```python
query =input("Enter your query : ")

# Convert to lowercase
query = query.lower()

# Remove punctuation (basic cleaning)
query = query.replace(".", "").replace(",", "")

words = query.split()

#store word frequency
frequency = {}

# Count word frequency
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Display only words repeated more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

## Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

#valid list
valid_readings = []

# Check for even values
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# domains and count
for email in emails:
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total = len(emails)

# Calculate the percentage usage of each domain
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {int(percentage)}%")
```

## Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
average = sum(sales) / len(sales)

# Detect spikes (>30% above average)
threshold = average * 1.3

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

## Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Count occurrences of user
for uid in user_ids:
    if uid in id_count:
        id_count[uid] += 1
    else:
        id_count[uid] = 1

# Display only duplicates
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
