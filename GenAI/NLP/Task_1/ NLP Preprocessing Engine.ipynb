# Task 1: Conceptual Understanding

# 1. Difference between "Love" and "love"
In NLP, "Love" and "love" are treated as different words due to case sensitivity.
This can lead to duplicate features in models. Converting text to lowercase helps
maintain consistency and improves model performance.

# 2. What happens if stopwords are not removed?
If stopwords are not removed, the dataset may contain unnecessary words like "is", "the", "and".
This increases noise and may reduce model efficiency and accuracy.

# 3. When removing stopwords can be harmful
a) Sentiment analysis:
   Example: "I am not happy" → removing "not" changes meaning completely.

b) Chatbots or question answering:
   Stopwords help maintain sentence structure and meaning.

# 4. Stemming vs Lemmatization
Stemming cuts words to base form without meaning (running → runn)
Lemmatization converts words to meaningful root form (running → run)

## Task 2:Build Advanced Preprocessing Function

```python
import re
from collections import Counter
def preprocess_text(text):
    
    # Handle empty input
    if not text or not isinstance(text, str):
        return [], ""
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    
    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)
    
    # Remove numbers
    text = re.sub(r"\d+", "", text)
    
    # Remove emojis and special characters
    text = re.sub(r"[^a-z\s]", "", text)
    
    # Handle repeated characters (soooo → soo)
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)
    
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    
    # Tokenization
    tokens = text.split()
    
    # Remove very short words (except no, not)
    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            cleaned_tokens.append(word)
    
    # Final cleaned sentence
    cleaned_sentence = " ".join(cleaned_tokens)
    
    return cleaned_tokens, cleaned_sentence
```

## Task 3: Stress Testing

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
results = []

for text in sentences:
    tokens, clean = preprocess_text(text)
    
    print("Original Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-" * 50)
    
    results.append(tokens)
```

## Task 4: Token Analytics

```python
for i, tokens in enumerate(results):
    
    if len(tokens) == 0:
        continue
    
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total
    
    print(f"Sentence {i+1}")
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Token Length:", round(avg_len, 2))
    print("-" * 40)
```

## Analysis Questions
# Which sentence had the most noise?
Likely the one with emojis, repeated characters, and symbols.

# Which sentence retained the most meaningful tokens after cleaning?

## Task 5: Frequency Analysis

```python
all_tokens = []

for tokens in results:
    all_tokens.extend(tokens)

freq = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(freq.most_common(10))

print("\nTop 5 least frequent words:")
print(freq.most_common()[-5:])
```

## Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    # Process each sentence
    for text in text_list:
        tokens, clean = preprocess_text(text)
        
        # Collect tokens and cleaned sentences
        all_tokens.extend(tokens)
        clean_sentences.append(clean)
    
    # Return final structured output
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
    
output = full_pipeline(sentences)

# Proper JSON-style output
import json
print(json.dumps(output, indent=4))
```
