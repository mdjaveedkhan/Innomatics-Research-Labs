## Problem Statement 1: Smart Parking Lot Management System

```python
# Function to manage parking lot
def manage_parking(capacity, logs):
    
    current = 0
    peak = 0
    
    # Loop through vehicle logs
    for entry in logs:
        
        if entry == "IN":
            current += 1
        elif entry == "OUT":
            current -= 1
        
        # Track peak usage
        if current > peak:
            peak = current
    
    print("Currently Parked Vehicles:", current)
    
    # Capacity check
    if current > capacity:
        print("Parking Status: Full - Alert!")
    else:
        print("Parking Status: Available")

capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking(capacity, logs)
```

## Problem Statement 2:Online Food Delivery Time Estimator

```python
# Function to estimate delivery time
def estimate_delivery(distance, traffic, weather):
    
    # Base time = 5 minutes per km
    time = distance * 5
    
    # Traffic level
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5
    
    # Weather comdition
    if weather == "Rainy":
        time += 5
    elif weather == "Storm":
        time += 15
    print("Estimated Delivery Time:", time, "minutes")

estimate_delivery(8, "High", "Rainy")
```

## Problem Statement 3:Movie Theatre Seat Occupancy

```python
# Function to analyze seat booking data
def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    # Calculating occupancy percentage
    percentage = (booked_count / total_seats) * 100
    print("Occupancy:", int(percentage), "%")
    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

# Example (150 seats booked out of 200)
booked = [1] * 150
analyze_occupancy(200, booked)
```

## Problem Statement 4:Cloud Server Load Classification System

```python
# Function to classify server load
def classify_server_load(readings):
    total = 0
    
    # Calculating average
    for value in readings:
        total += value
    average = total / len(readings)
    print("Average CPU Load:", int(average), "%")
    
    # Rules 
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

cpu_readings = [45, 60, 70, 85, 90]
classify_server_load(cpu_readings)
```

## Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
# Function that tracks usage of classroom resources 
def monitor_resources(resources):
    overused = []
    
    # Assume overuse if usage > 8 hours
    for item, hours in resources.items():
        if hours > 8:
            overused.append(item)
    if len(overused) > 0:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_resources(usage)
```

## Problem Statement 6: Online Event Registration Capacity Controller

```python
# Function to manage registrations
def manage_event(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        print("Confirmed Registrations:", confirmed)
        print("Waitlisted Users:", waitlisted)
        print("Registration Status: Closed")  
    else:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")

manage_event(100, 105)
```
